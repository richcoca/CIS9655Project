```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#add csv file to dataframe
#make sure that the code and csv file are in the same location
file ='clean_spd.csv'
df = pd.read_csv(file, parse_dates=True)
df.head()
#df = df["Election Year"].astype(object)
```

```python
df_year = df.groupby(['Election Year', 'Party'], as_index=False)['Amount'].sum()
df_year
```

```python
#Graph of percentage of total donation given to each party by year 
#in time series probably line graph
```

```python
df_republican=df_year.loc[df_year['Party'] == 'Republican']
df_republican

df_democrat=df_year.loc[df_year['Party'] == 'Democrat']
df_democrat

fig, ax = plt.subplots()
#republican 
ax.plot(df_republican["Election Year"], df_republican["Amount"], color = 'r')

#democrat
ax.plot(df_democrat["Election Year"], df_democrat["Amount"], color = 'b')

ax.set_xlabel("Election Year")
ax.set_ylabel("Amount Donated")
ax.set_title(" Percent of Donation Given to Each Party Over Time")

plt.show()
```
