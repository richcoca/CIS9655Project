```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sports= "SPD.csv"
sport= 'SPD(League only).csv'

df = pd.read_csv(sports,index_col=0, parse_dates=True)
dfd = pd.read_csv(sport)

dfd['colors']=0
dfd.loc[dfd['League']=='MLB', 'colors']='blue'
dfd.loc[dfd['League']=='NBA', 'colors']='red'
dfd.loc[dfd['League']=='NFL', 'colors']='green'
dfd.loc[dfd['League']=='NHL', 'colors']='yellow'
dfd.loc[dfd['League']=='NASCAR', 'colors']='pink'
dfd.loc[dfd['League']=='WNBA', 'colors']='purple'
dfd=dfd.sort_values(by=['League'])

dfd
```
