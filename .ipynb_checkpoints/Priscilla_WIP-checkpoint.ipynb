```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file ='spd.csv'
df = pd.read_csv(file, index_col=0, parse_dates=True)
df.head()
df.info()
df
```

```python
df1 = df.groupby(['Election Year', 'Party'], as_index=False)['Amount'].sum()
df1
```

```python
df12=df1.loc[df1['Party'] == 'Republican']
df12
```

```python
df123=df1.loc[df1['Party'] == 'Democrat']
df123
```

```python
df12=df1.loc[df1['Party'] == 'Republican']
df12

df123=df1.loc[df1['Party'] == 'Democrat']
df123

fig, ax = plt.subplots()
#republican 
ax.plot(df12["Election Year"], df12["Amount"], color = 'r')

#democrat
ax.plot(df123["Election Year"], df123["Amount"], color = 'b')

ax.set_xlabel("Election Year")
ax.set_ylabel("Amount Donated")
ax.set_title(" Percent of Donation Given to Each Party Over Time")

plt.show()
```
