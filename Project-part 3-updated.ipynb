```python
import pandas as pd

spd_data = pd.read_csv("SPD(League only).csv")

amount = spd_data.groupby(by=['State'])['Amount'].sum()
amount = amount.sort_values(ascending = False)

state = list(amount.index)
state = [i.strip()for i in state]
data = [list(z) for z in zip(state, list(amount))]
max_amount = max(amount)
```

```python
number = spd_data.groupby(by=['State'])['Amount'].count()
number = number.sort_values(ascending = False)

state_donate_number = list(number.index)
state_donate_number = [i.strip()for i in state_donate_number]
number_data = [list(z) for z in zip(state_donate_number, list(number))]
max_number = max(number)
print(number_data)
print(max_number)
```

```python
from pyecharts import options as opts
```

```python
from pyecharts.charts import Pie

pie = Pie().add("Donation Amount",
                data,
                radius = ["20%", "80%"],
                center = ["60%", "60%"],
                rosetype = "radius")

pie.set_global_opts(title_opts = opts.TitleOpts(title="Political Donation Amount by State",
                                                pos_right = '60%'),
                    legend_opts = opts.LegendOpts(
                                                orient='vertical',
                                                pos_right="75.7%",
                                                pos_top="20%"))
pie.set_series_opts(label_opts = opts.LabelOpts(formatter="{b} : {d}%"))    
pie.render_notebook()
```

```python
pie = Pie().add("# of Sport Owners donated",
                number_data,
                radius = ["20%", "80%"],
                center = ["60%", "60%"],
                rosetype = "radius")

pie.set_global_opts(title_opts = opts.TitleOpts(title="# of Sport Owners donates in each State",
                                                pos_right = '60%'),
                    legend_opts = opts.LegendOpts(
                                                orient='vertical',
                                                pos_right="77%",
                                                pos_top="20%"))
pie.set_series_opts(label_opts = opts.LabelOpts(formatter="{b} : {d}%"))    
pie.render_notebook()
```
